```python
def strassen(n):
    print(n)
    ans = []
    for entry in n:
        ans.append(7 * strassen(math.ceil(entry/2) + (9*(entry**2) / 2)))
    return ans

def conventional(n):
    ans = []
    for entry in n: 
        ans.append(2 * entry ** 3 - entry ** 2)
    return ans
```

```python
from numpy import *
import math
import matplotlib.pyplot as plt

t = linspace(1, 500, 400, dtype=int)

print(t)

a = strassen(t)
b = conventional(t)

plt.plot(t, a, 'r') # plotting t, a separately 
plt.plot(t, b, 'b') # plotting t, b separately 
plt.show()
```
